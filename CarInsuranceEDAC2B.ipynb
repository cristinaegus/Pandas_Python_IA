```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("govindaramsriram/car-insurance-premium-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(path + "/car_insurance_premium_dataset.csv")
df
```

```python
# listar las columnas
df.columns
```

```python
# crea un plot  con car manufacturin year y  Insurance Premiun
import matplotlib.pyplot as plt
def_20 = df.head(20)
def_20.plot.scatter(x="Car Manufacturing Year", y="Insurance Premium ($)")

plt.plot(df["Car Manufacturing Year"], df["Insurance Premium ($)"])
plt.xlabel('Car Manufacturing Year')
plt.ylabel('Insurance Premium ($)')
plt.title('Insurance Premium vs. Car Manufacturing Year')
plt.show()
```

```python
# crear un grafico con tres valores relacionados,  friver experince, insurance premium , y car age
import matplotlib.pyplot as plt
df.plot.scatter(x="Driver Experience", y="Insurance Premium ($)")
df.plot.scatter(x="Driver Experience", y="Car Age")

plt.plot(df["Driver Experience"], df["Insurance Premium ($)"], marker='x', linestyle='-', markersize=8)
plt.plot(df["Driver Experience"], df["Car Age"], marker='o', linestyle='-', markersize=6, color='red') # Example of changing color
plt.xlabel('Driver Experience')
plt.ylabel('Value') # Changed ylabel to be more general as we have two different values on y-axis
plt.title('Insurance Premium and Car Age vs. Driver Experience')
plt.legend(['Insurance Premium ($)', 'Car Age']) # Added a legend to distinguish the lines
plt.show()
```

```python
# haz los calculos solo con los 20 primeros conductore
df_20 = df.head(20)
df_20.plot.scatter(x="Driver Experience", y="Insurance Premium ($)")
```

Insurance Premium vs. Driver Experience: The blue line with markers shows the relationship between Driver Experience and Insurance Premium. It appears there isn't a strong linear relationship. The insurance premium seems to fluctuate as driver experience increases, and the markers help to see the distribution of premiums at different experience levels. There might be some clusters of points or outliers that are more apparent with the markers.
Car Age vs. Driver Experience: The orange line with markers shows the relationship between Driver Experience and Car Age. Similar to the insurance premium, there doesn't seem to be a clear linear trend between car age and driver experience. The markers show that there is a wide range of car ages for any given level of driver experience.
Adding the markers helps to visualize the spread of the data points and see how the individual values of Insurance Premium and Car Age are distributed across different levels of Driver Experience. It reinforces that these relationships are not simple linear ones and suggests that other factors might be influencing the insurance premium and car age.
